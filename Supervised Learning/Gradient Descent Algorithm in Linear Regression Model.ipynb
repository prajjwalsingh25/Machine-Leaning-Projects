```python
import numpy as np
import matplotlib.pyplot as plt
import math, copy
```

```python
# Generate 20 random data points for x_train
x_train = np.random.uniform(0, 10, size=20)

# Define parameters for introducing noise
noise_mean = 0
noise_std = 100

# Generate random noise with mean=0 and standard deviation=10
noise = np.random.normal(noise_mean, noise_std, size=20)

# Define the parameters for the linear equation
m = 50
b = 100

# Compute corresponding y_train values using the linear equation y = mx + b with added noise
y_train = m * x_train + b + noise

# Print the generated data points
print("Generated Data Points:")
print("x_train =", x_train)
print("y_train =", y_train)
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
    return f_wb 
```

We've observed that by manually adjusting the parameters of the linear regression model, we can bring the predicted values closer to the actual values. However, finding the exact values of these parameters that result in the closest match between actual and predicted values is challenging. To address this challenge, we propose to utilize the Gradient Descent Algorithm. This algorithm will iteratively adjust the parameters of the model in the direction that minimizes the difference between predicted and actual values, ultimately leading to a more optimal set of parameter values that yield accurate predictions.

# Cost Function

```python
def compute_cost(x, y, w, b): 
    # number of training examples
    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

# Gradient Descent Algorithm

```python
def compute_gradient(x, y, w, b): 
  # Number of training examples
    m = x.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):  
        f_wb = w * x[i] + b 
        dj_dw_i = (f_wb - y[i]) * x[i] 
        dj_db_i = f_wb - y[i] 
        dj_db += dj_db_i
        dj_dw += dj_dw_i 
    dj_dw = dj_dw / m 
    dj_db = dj_db / m 
        
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, gradient_function): 
    
    w = copy.deepcopy(w_in) # avoid modifying global w_in
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    p_history = []
    b = b_in
    w = w_in
    
    for i in range(num_iters):
        # Calculate the gradient and update the parameters using gradient_function
        dj_dw, dj_db = gradient_function(x, y, w , b)     

        # Update Parameters using equation (3) above
        b = b - alpha * dj_db                            
        w = w - alpha * dj_dw                            

        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( cost_function(x, y, w , b))
            p_history.append([w,b])
        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters/10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
 
    return w, b, J_history, p_history #return w and J,w history for graphing
```

```python
# initialize parameters
w_init = 0
b_init = 0
# some gradient descent settings
iterations = 1000
tmp_alpha = 1.0e-2
# run gradient descent
w_final, b_final, J_hist, p_hist = gradient_descent(x_train ,y_train, w_init, b_init, tmp_alpha, 
                                                    iterations, compute_cost, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

After undergoing 1000 iterations of gradient descent the optimum values of w and b comes out to be 65.7221, 11.4814 respectively

As the iterations progress, both the change in w (dj_dw) and b (dj_db) become smaller, approaching zero. This suggests that the algorithm is converging towards an optimal solution.

he parameters w and b stabilize around certain values after a certain number of iterations. In this case, w stabilizes around 65.7221 and b around 11.4814, indicating that these values are close to the optimal parameters for the linear regression model
